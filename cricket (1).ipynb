```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/T20 data')
```

remove missing value

```python
df['bowler2_id'].fillna(df.groupby(['bowler1_id'])['bowler2_id'].transform('mean'), inplace=True)
df['bowler2_name'].fillna(df.groupby(['bowler1_name'])['bowler2_name'].transform(lambda x: x.mode().iloc[0]), inplace=True)
df['bowler2_overs'].fillna(df.groupby(['bowler1_name'])['bowler2_overs'].transform('median'), inplace=True)
df['bowler2_runs'].fillna(df.groupby(['bowler1_name'])['bowler2_runs'].transform('median'), inplace=True)
print(df)
```

EDA

```python
best_team = df.groupby('current_innings')['runs'].sum().reset_index()

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(best_team['current_innings'], best_team['runs'], color='skyblue')
plt.xlabel('Teams')
plt.ylabel('Total Runs')
plt.title('Total Runs scored by Inning Team')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
df['combined'] = df['batsman1_name'] + ' ' + df['batsman1_runs'].astype(str)

# Calculate the total combined runs
total_combined_runs = df['combined'].sum()

# Plot the graph
plt.figure(figsize=(10, 6))
plt.bar(df['batsman1_name'], df['batsman1_runs'], color='skyblue')
plt.xlabel('Batsman Name')
plt.ylabel('Runs')
plt.title('Runs Scored by Each Batsman')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
first_match_details = df.iloc[1]
print("Details of the First Match:")
print(first_match_details)
```

```python
balls_per_inning = df.groupby('innings_id').size()

# Plotting graph
plt.figure(figsize=(10, 6))
balls_per_inning.plot(kind='bar', color='skyblue')
plt.xlabel('Innings ID')
plt.ylabel('Number of Balls Bowled')
plt.title('Number of Balls Bowled in Each Inning')
plt.grid(True)
plt.tight_layout()
plt.show()

# Displaying total balls bowled
print("Total Balls Bowled in Each Inning:\n", balls_per_inning)
```

```python
top_10_batsmen = df.sort_values(by='runs', ascending=False).head(10)

# Plot graph
plt.figure(figsize=(10, 6))
plt.bar(top_10_batsmen['batsman1_name'], top_10_batsmen['runs'], color='skyblue')
plt.xlabel('Batsman Name')
plt.ylabel('Runs Scored')
plt.title('Top 10 Batsmen by Runs Scored')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
top_bowlers = df.groupby('bowler1_name')['over'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
plt.bar(top_bowlers.index, top_bowlers.values, color='skyblue')
plt.xlabel('Bowler Name')
plt.ylabel('Total Overs Bowled')
plt.title('Top 10 Bowlers by Total Overs Bowled')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Displaying top bowlers
print("Top 10 Bowlers by Total Overs Bowled:\n", top_bowlers)
```

```python
inning1_summary = df.groupby(['innings_id']).sum()[['isWide', 'isBoundary', 'isNoball']]

# Plot graph
fig, axes = plt.subplots(3, 1, figsize=(5,6), sharex=True)

# Plot for wide
inning1_summary['isWide'].plot(kind='bar', ax=axes[0], color='r')
axes[0].set_title('iswide')
axes[0].set_ylabel('Count')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Plot for boundary
inning1_summary['isBoundary'].plot(kind='bar', ax=axes[1], color='g')
axes[1].set_title('isboundary')
axes[1].set_ylabel('Count')
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

# Plot for no ball
inning1_summary['isNoball'].plot(kind='bar', ax=axes[2], color='b')
axes[2].set_title('No Ball')
axes[2].set_ylabel('Count')
axes[2].grid(axis='y', linestyle='--', alpha=0.7)

plt.xlabel('Innings ID')
plt.tight_layout()
plt.show()
```
